```python
import pandas as pd

# Load your cleaned dataset
df = pd.read_csv("clean_drug_reviews.csv")

print("Original shape:", df.shape)
print(df["Rating"].value_counts().sort_index())

# -----------------------------------
# 1. Standard binary dataset
# negative = 1-4
# positive = 7-10
# drop 5-6
# -----------------------------------
binary_standard = df[(df["Rating"] <= 4) | (df["Rating"] >= 7)].copy()

def label_standard(r):
    if r <= 4:
        return 0   # negative
    else:
        return 1   # positive

binary_standard["label"] = binary_standard["Rating"].apply(label_standard)

print("\nStandard binary shape:", binary_standard.shape)
print(binary_standard["label"].value_counts())

binary_standard.to_csv("binary_standard.csv", index=False)
```
