```python
import pandas as pd

# Load your cleaned dataset
df = pd.read_csv("clean_drug_reviews.csv")

print("Original shape:", df.shape)
print(df["Rating"].value_counts().sort_index())

# -----------------------------------
# 3-class dataset
# negative = 1-4
# neutral  = 5-6
# positive = 7-10
# -----------------------------------
three_class_df = df.copy()

def label_three_class(r):
    if r <= 4:
        return 0   # negative
    elif r <= 6:
        return 1   # neutral
    else:
        return 2   # positive

three_class_df["label"] = three_class_df["Rating"].apply(label_three_class)

print("\n3-class dataset shape:", three_class_df.shape)
print(three_class_df["label"].value_counts())

three_class_df.to_csv("three_class_dataset.csv", index=False)

print("\nSaved: three_class_dataset.csv")
```
