```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("clean_drug_reviews.csv")
print("Original cleaned dataset shape:", df.shape)
print(df["label"].value_counts())
```

```python
df_neg = df[df["label"] == 0].copy()
df_neu = df[df["label"] == 1].copy()
df_pos = df[df["label"] == 2].copy()

print("\nNegative count:", len(df_neg))
print("Neutral count :", len(df_neu))
print("Positive count:", len(df_pos))
```

```python
min_class = min(len(df_neg), len(df_neu), len(df_pos))

balanced_df = pd.concat([
    df_neg.sample(min_class, random_state=42),
    df_neu.sample(min_class, random_state=42),
    df_pos.sample(min_class, random_state=42)
]).sample(frac=1, random_state=42).reset_index(drop=True)

print("\nBalanced dataset shape:", balanced_df.shape)
print(balanced_df["label"].value_counts())
```

```python
pos_dom_df = pd.concat([
    df_neg.sample(10000, random_state=42),
    df_neu.sample(5000, random_state=42),
    df_pos.sample(20000, random_state=42)
]).sample(frac=1, random_state=42).reset_index(drop=True)

print("\nPositive-dominant dataset shape:", pos_dom_df.shape)
print(pos_dom_df["label"].value_counts())
```

```python

neg_dom_df = pd.concat([
    df_neg.sample(20000, random_state=42),
    df_neu.sample(5000, random_state=42),
    df_pos.sample(10000, random_state=42)
]).sample(frac=1, random_state=42).reset_index(drop=True)

print("\nNegative-dominant dataset shape:", neg_dom_df.shape)
print(neg_dom_df["label"].value_counts())
```

```python
balanced_df.to_csv("balanced_dataset.csv", index=False)
pos_dom_df.to_csv("positive_dominant_dataset.csv", index=False)
neg_dom_df.to_csv("negative_dominant_dataset.csv", index=False)

print("\nDatasets saved successfully:")
print(" - balanced_dataset.csv")
print(" - positive_dominant_dataset.csv")
print(" - negative_dominant_dataset.csv")
```
